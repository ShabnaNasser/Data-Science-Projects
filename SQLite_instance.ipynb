Q10: Produce a list of facilities with a total revenue less than 1000. 

The output of facility name and total revenue, sorted by revenue. Remember 

that there's a different cost for guests and members! 

```python
c.execute("SELECT name, \
                                SUM(CASE when memid = 0 then slots * guestcost else slots * membercost END) AS revenue \
                                FROM bookings \
                                LEFT JOIN facilities \
                                USING(facid) \
                                GROUP BY name \
                                HAVING revenue < 1000;")
```

```python
pd.read_sql_query("SELECT name,SUM(CASE when memid = 0 then slots * guestcost else slots * membercost END) AS revenue FROM bookings LEFT JOIN facilities USING(facid) GROUP BY name HAVING revenue < 1000;", conn)
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
pd.read_sql_query("SELECT a.surname || ', ' || a.firstname AS members, \
                            b.surname || ', ' || b.firstname AS recommended_by \
                            FROM members a, members b \
                            WHERE a.recommendedby >0 AND a.recommendedby = b.memid \
                            ORDER BY b.surname;", conn)
```

Q12: Find the facilities with their usage by member, but not guests

```python
pd.read_sql_query("SELECT name, firstname || ' ' || surname AS member_name, \
                            COUNT(surname) AS 'usage' \
                            FROM bookings \
                            LEFT JOIN facilities USING(facid) \
                            LEFT JOIN members USING(memid) \
                            WHERE memid != 0 \
                            GROUP BY name, member_name;", conn)
```

Q13: Find the facilities usage by month, but not guests 

```python
pd.read_sql_query("SELECT strftime('%m', starttime) AS month, name, COUNT(name) AS 'usage' \
                            FROM bookings \
                            LEFT JOIN facilities USING(facid)\
                            WHERE memid != 0 \
                            GROUP BY month, name;",conn)
```
